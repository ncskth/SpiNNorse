```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
'''
This function creates a train of spikes
'''
def create_spikes(nb_spikes, nb_steps, random):  
    
    i_spikes = np.zeros(nb_steps)
    if random:
        for i in np.random.randint(nb_steps, size=nb_spikes):
            i_spikes[i] = 1
    else:    
        tbs = math.ceil(nb_steps/(nb_spikes+1)) 
        next_spike = tbs
        for i in range(0,int(nb_steps)):
            if i == next_spike:
                i_spikes[i] = 1
                next_spike = next_spike + tbs
        i_spikes = np.array(i_spikes)
    i_indexes = np.where(i_spikes>0)
    
    return i_spikes, i_indexes
```

```python
path = "summary/common/"

nb_steps = 200 # total number of steps

min_nb_spikes = 10
max_nb_spikes = 80
nb_trains = 12

nb_spikes = range(min_nb_spikes,max_nb_spikes+1,math.ceil((max_nb_spikes-min_nb_spikes)/(nb_trains)))
i = 0
for nb_spk in nb_spikes[0:nb_trains]:
    i += 1
    opt_rnd = 1
    i_spikes, i_indexes = create_spikes(nb_spk, nb_steps, opt_rnd)
    simname = ("%02d" % (i,))
    np.savetxt(path + "spikes_" + simname + ".csv", i_spikes, delimiter=",")

        
```
